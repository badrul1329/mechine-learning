**GRADENT DESCENT for Linear Regression**

```python
m=0
c=0
L=.0001
iterations=1000
n=len(X) # total no of sample of X

for a in range(iterations):
  y_pred = m*X + c
  deri_m = (-2/n)*sum(X*(y - y_pred))
  deri_c = (-2/n)*sum(y - y_pred)

  #calculate the value of  m and c
  m= m - (L * deri_m)
  c= c - (L * deri_c)
  print (m,c)

```

```python
#Last iteration
print (m,c)
```
